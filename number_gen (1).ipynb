```python
from random import randint
import os.path
import phonenumbers

# Function to generate a valid US phone number using the template
def generate_us_phone_number(area_code):
    number = ''.join(str(randint(0, 9)) for _ in range(7))  # Generate a 7-digit number
    return f"({area_code}) {number[:3]}-{number[3:]}"

# Load the area codes from the template file
template_file_path = '/Users/henry/Documents/num_gen/USA.txt'
with open(template_file_path, 'r') as template_file:
    area_codes_template = [line.split('|')[0].strip() for line in template_file]

# Get the area code from the user
desired_area_code = input("Enter the desired area code: ")

# Validate if the desired area code is in the template
if desired_area_code in area_codes_template:
    save_to_file = input("Save validated numbers to a file? [y/n]: ").lower()
    if save_to_file == "y":
        output_file_name = input("Enter the name of the output file for validated numbers: ")
        output_file_path = f'{output_file_name}.txt'
        
        n = int(input(f"How many phone numbers to generate for area code {desired_area_code}: "))
        validated_numbers = []

        for _ in range(n):
            phone_number = generate_us_phone_number(desired_area_code)
            
            # Validate the generated number using phonenumbers library
            parsed_number = phonenumbers.parse(phone_number, 'US')  # Specify the default region as 'US'
            if phonenumbers.is_valid_number(parsed_number):
                validated_numbers.append(phone_number)
        
        # Write validated numbers to the specified output file
        with open(output_file_path, 'w') as output_file:
            output_file.write('\n'.join(validated_numbers))
        
        print(f"Generation and validation complete. Validated numbers saved to '{output_file_path}'")
    else:
        n = int(input(f"How many phone numbers to generate for area code {desired_area_code}: "))
        validated_numbers = []

        for _ in range(n):
            phone_number = generate_us_phone_number(desired_area_code)
            
            # Validate the generated number using phonenumbers library
            parsed_number = phonenumbers.parse(phone_number, 'US')  # Specify the default region as 'US'
            if phonenumbers.is_valid_number(parsed_number):
                validated_numbers.append(phone_number)
                print(f"Generated and Validated: {phone_number}")
            else:
                print(f"Generated but Invalid: {phone_number}")

        print("Generation and validation complete. Validated numbers not saved.")
else:
    print(f"The desired area code '{desired_area_code}' is not in the template.")
```
